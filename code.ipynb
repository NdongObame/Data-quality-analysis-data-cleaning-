```python
def missingv(a):
    for x in a.columns:
        print ( x + ' : ' + str(a[x].isna().sum()))
```

```python
def unique(a):
    for x in a.columns:
        if "date" not in x :
            print( x + ":\n" + str(a[x].unique()) +  "\n" + str(a[x].dtype) + "\n")
```
